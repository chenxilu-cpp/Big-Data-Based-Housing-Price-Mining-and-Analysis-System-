```python
import pandas as pd
from pyecharts.charts import Pie, Line, Bar
from pyecharts import options as opts

from pyecharts.globals import CurrentConfig, NotebookType

# 设置 Jupyter Notebook 环境
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_NOTEBOOK

# 加载数据
df = pd.read_csv('data.csv')

# 数据预处理：去除空格和转换数据类型
df['售价'] = df['售价'].astype(float)
df['单价'] = df['单价'].astype(float)
df['面积'] = df['面积'].astype(float)
df['楼层'] = df['楼层'].str.extract('(\d+)').astype(float)  # 提取楼层数字部分
#如果年份只年份的信息，如‘2014’就不用处理，若年份列包含"2014年"格式的信息，我们首先去掉非数字字符
df['年份'] = df['年份'].str.extract('(\d{4})')  # 提取四位数年份
df['年份'] = pd.to_datetime(df['年份'], format='%Y', errors='coerce').dt.year

# 检查建筑类型是否存在缺失值，并处理缺失值（这里用最常见的建筑类型填充）
if df['建筑类型'].isnull().sum() > 0:
    most_common_type = df['建筑类型'].mode()[0]  # 找到最常见的建筑类型
    df['建筑类型'].fillna(most_common_type, inplace=True)
# 检查年份是否存在缺失值，并处理缺失值（这里用中位数填充）
if df['年份'].isnull().sum() > 0:
    median_year = df['年份'].median()  # 找到年份的中位数
    df['年份'].fillna(median_year, inplace=True)
# 查看数据前几行'
df.head()
```

```python
import pandas as pd
from pyecharts.charts import Line
from pyecharts import options as opts

# 加载数据
df = pd.read_csv('data.csv')
df.head()

df['年份'] = df['年份'].str.extract('(\d{4})')  # 提取四位数年份
df['年份'] = pd.to_datetime(df['年份'], format='%Y', errors='coerce').dt.year

# 计算每年平均房价（假设有一个名为 '房价' 的列）
yearly_price = df.groupby('年份')['售价'].mean().reset_index()

# 时序分析：绘制房价随年份的变化趋势
line = (
    Line()
    .add_xaxis(yearly_price['年份'].tolist())
    .add_yaxis("平均房价", yearly_price['售价'].tolist(), is_smooth=True)
    .set_global_opts(
        title_opts=opts.TitleOpts(title="房价随年份变化趋势"),
        xaxis_opts=opts.AxisOpts(type_="category"),
        yaxis_opts=opts.AxisOpts(type_="value"),
        tooltip_opts=opts.TooltipOpts(trigger="axis"),
    )
)
line.render_notebook()
```

```python
# 计算各大区域的平均单价
avg_price_by_area = df.groupby('区域')['单价'].mean().sort_values(ascending=False)

# 获取最贵和最便宜的区域
most_expensive_area = avg_price_by_area.idxmax()
cheapest_area = avg_price_by_area.idxmin()

# 打印最贵和最便宜的区域
print(f"最贵的区域：{most_expensive_area}，平均单价：{avg_price_by_area.max()}")
print(f"最便宜的区域：{cheapest_area}，平均单价：{avg_price_by_area.min()}")
```

```python
# 导入需要的库
from pyecharts.charts import Pie

# 准备数据
avg_price_data = avg_price_by_area.reset_index()
area_names = avg_price_data['区域'].tolist()
avg_prices = avg_price_data['单价'].tolist()

# 创建饼状图
avg_price_pie = (
    Pie()
    .add(
        "",
        [list(z) for z in zip(area_names, avg_prices)],
        center=["50%", "50%"],
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="各大区域平均单价"),
        legend_opts=opts.LegendOpts(type_="scroll", pos_left="80%", orient="vertical"),
    )
    .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))
)

# 显示饼状图
avg_price_pie.render_notebook()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA
import warnings

warnings.filterwarnings('ignore')


yearly_avg_price = df.groupby('年份')['单价'].mean().reset_index()
yearly_avg_price.set_index('年份', inplace=True)

# 构建ARIMA模型
model = ARIMA(yearly_avg_price['单价'], order=(5, 1, 0))
model_fit = model.fit()

# 预测未来的房价
forecast_steps = 10
forecast = model_fit.forecast(steps=forecast_steps)
print(forecast)

# 将年份索引转换为整数
last_year = int(yearly_avg_price.index[-1])

# 创建预测年份序列
predicted_years = np.arange(last_year + 1, last_year + 1 + forecast_steps)

# 将预测结果转换为 Series
predicted_series = pd.Series(forecast, index=predicted_years)

# 结合实际数据和预测数据
combined_series = pd.concat([yearly_avg_price['单价'], predicted_series])

# 可视化
plt.figure(figsize=(10, 6))
plt.plot(combined_series.index, combined_series.values, label='Actual and Predicted')
plt.axvline(x=yearly_avg_price.index[-1], color='r', linestyle='--', label='Forecast Start')
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.title('House Price Trend and Forecast')
plt.legend()
plt.show()
```
